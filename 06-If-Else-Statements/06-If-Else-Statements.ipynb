# 06-If-Else-Statements


Conditional statements, often referred to as if-then statements, let you control what pieces of code are run based on the value of some Boolean condition. Here's an example:

```python

    if age >= 16:                    
       print("You can drive in US!") 
    elif age >= 18:                  
        print("You can drive!")
    else:                            
        print("Be health, ride a bike!")
```

Basically you can use if statements to controll the flow of execution.

### Logic Operators

In this module you will recap the logic operators to use in if statements.

- Equals: a == b
- Not Equals: a != b
- Less than: a < b
- Less than or equal to: a <= b
- Greater than: a > b
- Greater than or equal to: a >= b

### Indentation

Identation is very important in Python. With identation, is possible to delimit the scope of execution of your code, so in every python code, the identation is checked:

```python
if age > 18:
print("You got an error!") # this code is not idented

#Right way

if age > 18:
    print("OK")            # this code is idented

```


### Short hand If and If-Else (Ternary Operators)

You can write your if statements "inline". 

This form is called "Ternary Operators", and can be used in simple expressions:

```python
a = 2
b = 330

# If
if a > b: print("a is greater than b")
    
# If-Else
print("A") if a > b else print("B")
```

### Nested If

Its possible to write nested Ifs.

```python
coins = 1000
name = "Mario"

if coins >= 1000:
    if name == "Mario":
        print("You can enter in the castle sir!")
    else:
        print("Please, wait here!")
else:
    print("Stay away or I will call the police!")

```

### in - not in

The operator "in" and "not" can be used to check items inside a list:

```python

pizza_ingredients = ["pineapple","chocolate","salt","cherry"]

if "pineapple" in pizza_ingredients:
    print("Yuummmyy :P")
    
if "pepperoni" not in pizza_ingredients:
    print("can I call this \"pizza\"?")

```


### and

To create more complex and easy to read boolean sentences, use the operator "and":

```python
pizza_ingredients = ["pineapple","chocolate","salt","cherry"]

if "pineapple" in pizza_ingredients and "pepperoni" not in pizza_ingredients :
    print("really?")
     
```

### or

You can use the operator "or" too:

```python
pizza_ingredients = ["pineapple","chocolate","salt","cherry"]

if "pineapple" in pizza_ingredients or "pepperoni" in pizza_ingredients :
    print("Yuummmy")
     
```
