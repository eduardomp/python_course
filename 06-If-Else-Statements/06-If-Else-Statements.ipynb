# 06-If-Else-Statements


Conditional statements, often referred to as if-then statements, let you control what pieces of code are run based on the value of some Boolean condition. Here's an example:

```python

    if age >= 16:                    
       print("You can drive in US!") 
    elif age >= 18:                  
        print("You can drive!")
    else:                            
        print("Be health, ride a bike!")
```

Basically you can use if statements to controll the flow of execution.

### Logic Operators

In this module you will recap the logic operators to use in if statements.

- Equals: a == b
- Not Equals: a != b
- Less than: a < b
- Less than or equal to: a <= b
- Greater than: a > b
- Greater than or equal to: a >= b

### Indentation

Identation is very important in Python. With identation, is possible to delimit the scope of execution of your code, so in every python code, the identation is checked:

```python
if age > 18:
print("You got an error!") # this code is not idented

#Right way

if age > 18:
    print("OK")            # this code is idented

```


### Short hand If and If-Else (Ternary Operators)

You can write your if statements "inline". 

This form is called "Ternary Operators", and can be used in simple expressions:

```python
a = 2
b = 330

# If
if a > b: print("a is greater than b")
    
# If-Else
print("A") if a > b else print("B")
```
